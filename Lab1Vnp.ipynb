```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/" 
```

```python
response = requests.get(url)
```

```python
raw_html = response.text
soup = BeautifulSoup(raw_html, "html.parser")
```

```python
def TabelaProductFunk(products):
    GotoviProdukti = []
    for product in products:
        ProductTitle = product.select_one('.wd-entities-title a').text.strip()
        Price = product.select('.woocommerce-Price-amount bdi')
        productURL = product.select_one('.product-image-link').get("href")
        CartButton = product.select_one('.wd-add-btn.wd-add-btn-replace a').get("href")
        
        if len(Price) > 1:
            regular_price = Price[0].text.strip().replace('\xa0', ' ')
            discount_price = Price[1].text.strip().replace('\xa0', ' ')
        else:
            regular_price = Price[0].text.strip().replace('\xa0', ' ')
            discount_price = "Nema popust"
            
        product_dict = {
            "ProductName": ProductTitle,
            "RegularPrice": regular_price,
            "DiscountPrice": discount_price,
            "ProductURL" : productURL,
            "AddToCartButton" : CartButton
        }
        
        GotoviProdukti.append(product_dict)
    
    return GotoviProdukti
```

```python
# posle / vnesuvame broj za strana
base_url = "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/"
all_products = []
# ima vkupno 14 strani
for i in range(1,14): 
    fullURL = base_url + str(i)
    response = requests.get(fullURL)
    soup = BeautifulSoup(response.text,"html.parser")
    products = soup.select('.product-wrapper')
    result=TabelaProductFunk(products)
    all_products=all_products+result
```

```python
PandaTabela = pd.DataFrame(all_products)
PandaTabela
```
